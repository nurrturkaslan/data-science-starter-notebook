# birleştirme(join) işlemleri

```python
import numpy as np
import pandas as pd
n = np.random.randint(1,30, size = (5,3))
df1 = pd.DataFrame(n, columns = ["var1","var2","var3"])
df1
```

```python
df2 = df1 + 99
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], ignore_index=False)
```

```python
pd.concat([df1,df2], ignore_index=True)
```
