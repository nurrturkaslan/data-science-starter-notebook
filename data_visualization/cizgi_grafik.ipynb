# cizgi grafik

```python
import seaborn as sns
fmri = sns.load_dataset("fmri")
df = fmri.copy()
df.head()
```

```python
df.shape
```

```python
# amaç her bir timepoint e göre signal gözlemlemek
```

```python
df["timepoint"].describe()
```

```python
df["signal"].describe()
```

```python
df.groupby("timepoint")["signal"].count()
```

```python
df.groupby("signal").count()
```

```python
df.groupby("timepoint")["signal"].describe()
```

```python
# çizgi garfik oluşturulması line plot
```

```python
sns.lineplot(x = "timepoint", y = "signal",data = df);
```

```python
sns.lineplot(x = "timepoint", y = "signal", hue = "event", data = df);
```

```python
sns.lineplot(x = "timepoint", y = "signal", hue = "event", style= "event",data = df);
```

```python
sns.lineplot(x = "timepoint",
             y = "signal",
             hue = "event",
             style = "event",
             markers = True,
             dashes = False,
             data = df);
#ortalama değerleri işaretlemek
```

```python
sns.lineplot(x = "timepoint",
             y = "signal",
             hue = "region",
             style = "event",
             markers = True,
             dashes = False,
             data = df);
```
