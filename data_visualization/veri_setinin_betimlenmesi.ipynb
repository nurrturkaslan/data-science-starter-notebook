# veri setinin betimlenmesi

```python
import seaborn as sns
planets = sns.load_dataset("planets")
df = planets.copy()
```

```python
df.head()
```

```python
df.shape #gözlem sayısı ve değişkene erişmek istersek
```

```python
df.columns #sadece değişkene erişmek istersek
```

```python
df.describe().T #veri setinin içerisinde yer alan betimsel istatistiklere erişelim fakat bunun yerine
#transpozunu alırsak bize daha güzel bir bilgi sunar 
#bu eksik gözlemleri göz ardı eder ve kategorik değişkenleri dışarıda bırakır
#yani buradaki tüm değişkenler sayısal değişkendir
#tabii bu şu anlama gelmiyor buradaki tüm sayısal değişkenler kategorik değişken olamaz 
#anlamına gelmiyor
```

```python
df.describe(include = "all").T
#inculude diyerek tüm değişkenleri barındırmasını isteyebilirz
#fakat çok anlamlı olmaya bilir.
#daha sonra ayrı ayrı ele alacağız
```
