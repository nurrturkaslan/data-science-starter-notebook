# ÇAPRAZLAMALAR

```python
import seaborn as sns
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
df.head()
```

```python
# bize sürekli değişkenin bir grafiğini oluştur dediğimiz de seaborn bize ;
sns.kdeplot(df.price, shade = True);
```

```python
#grafiğin bizden gizlediği mavi alanı oluşturan nedir?
```

```python
#facetgrid garfik üzerine eklenen boyutları bölerek göstermek için kullanılır
#hue : boyut ekleme
#height : yükseklik parametre
#map : eşleme yani maplememiz lazım
#add legend : bilgi ekleme 

```

```python
(sns
 .FacetGrid(df,
               hue = "cut",
               height = 5,
               xlim = (0, 10000))
.map(sns.kdeplot, "price", shade = True)
.add_legend()
);
```

```python
#price değişkeninin üzerine cut kategorik deg. ekledik
```

```python
sns.catplot(x = "cut",
           y = "price",
           hue = "color",
           kind = "point",
           data = df);
```

```python
#kind : tür point: noktalar ile ifade
#ayırt edicilik: diğer değişkenlerin price içinde ayırt edici bilgi taşıması 
#oluyot.
#ideal sınıfın altında color kat. degis. farklı sııflarının farklı degerler olusturmasının 
#burada colorun cut ile ile kesiştirildiğinde ideal sınıf altında ayırt edici bilgi taşığı anlamına geliyor
#price değişkenine göre
```
