# sürekli değişken özetleri

```python
import seaborn as sns
planets = sns.load_dataset("planets")
df = planets.copy()
df.head()
```

```python
df_num = df.select_dtypes(include = ["float64","int64"])
df_num.head()
#include dahil etmek demek
#asagıda sadece sürekli değişkenleri seçmiş olduk
```

```python
#sürekli değişkenlerin üzeerinde betimsel istatistiklere erimek için
#describe fonk kullanıyoruz
df_num.describe().T
```

```python
#değişkenlerin üzerindede betimlere erişmek istersek
df_num["distance"].describe()
```

```python
print("Ortalama: " + str(df_num["distance"].mean()))
print("Dolu Gözlem Sayısı: " + str(df_num["distance"].count()))
print("Maksimum Değeri: " + str(df_num["distance"].max()))
print("Minimum Değeri: " + str(df_num["distance"].min()))
print("Medyan: " + str(df_num["distance"].median()))
print("Standart Sapma: " + str(df_num["distance"].std()))
```
