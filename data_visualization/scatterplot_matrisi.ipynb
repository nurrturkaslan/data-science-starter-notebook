```python
import seaborn as sns
iris = sns.load_dataset("iris")
df = iris.copy()
df.head()
```

```python
df.shape
```

```python
sns.pairplot(df);
```

```python
# saçılım grafiği görsellerinde eğer oluşturulmış olan grafik bir toz bulutu şeklinde ise
#veya yapısal bir formu yoksa
# o iki değişken arasında bir ilişki olmadığı anlamına gelir
#gözlem noktaları incelendiğinde bunlar eğer birbirinden farklı köşelerde kümeleniyorsa
#bu durumda sık sık vurguladığımız çaprazlama durumundan kaynaklı olan oluşan bu yapıları yani 
#farklı farklı kümelenmeleri ifade eden bazı alt gruplar vardır
#bunların göz önünde bulundurulması gerekir
```

```python
#hemen bir kategorik değişkeni grafiğe boyut olarak ekleyerek inceleyelim
```
