# korelasyon grafiği oluşturulması

```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
# değişkenler arasındaki ilişkiyi ifade eden istatistiksel terimdir
#scatterplot:iki değişkenin arasındaki ilişki saçılım grafiği açılımıdır. 
#sayısal değişkenler arası ilişkiyi gösterir
#bu iki değişken bir arada değerlendiğinde veri seti içinde çıkan yapı hakkında bilgi verir.
# bu değişken kategorik ise sütun grafiği kullandık barplot
#bsayısal ise sürekli/kesikli ise histogram yoğunluk boxplot ve violin
#iki sayısal değişkenli veri görselleştirme yapcaz
```

```python
sns.scatterplot(x = "total_bill", y ="tip", data = df);
```

# korelasyon çaprazlamalar

```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "time", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "day",style = "day", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "size", size = "size", data = df);
```

# doğrusal ilişkinin gösterilmesi

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#lmplot lineer model kısaltılmışı doğrusal çizgi kullanacağımızda bunu kulanıyorum
sns.lmplot(x = "total_bill", y ="tip", data = df);
```

```python
sns.lmplot(x = "total_bill", y ="tip",hue = "smoker", data = df);
```

```python
sns.lmplot(x = "total_bill", y ="tip",hue = "smoker", col = "time", data = df);
```

```python
# col argümanı iki farklı grafiğe dönüştürdü
```

```python
sns.lmplot(x = "total_bill", y ="tip",hue = "smoker", col = "time",row = "sex", data = df);
```
